```python
results = []
class RiaSrapper(scrapy.Spider):
    #count = 0
    name = 'RiaMigrations'
    def __init__(self):
        options = webdriver.ChromeOptions()
        #options.headless = True
        #options.add_argument("headless")
        #desired_capabilities = options.to_capabilities()
        #self.driver = webdriver.Chrome(options = options) 
        
    def start_requests(self):
        url = 'https://www.garant.ru/news/tag/334/?ysclid=m3ee842w7g49890035'
        yield scrapy.Request(url = url,  headers = {'User-Agent': 'Mozilla/5.0' }, callback = self.parse_page)
    def parse_page(self, response):
        hrefs = response.css('div.descr>a::attr(href)').extract()
        for href in hrefs:
            yield response.follow(href, callback = self.parse_news)
            
    def parse_news(self, response):
        href = response.url
        title = response.css('div.page-content>h1::text').get()
        text = response.css('div.page-content>p::text').extract()
        date = response.css('div.actions-info>time::text').get()
        alt_title = title
        results.append({'href': response.url, 'title': title, 'alt_title': alt_title, 'date': date, 'text': ' '.join(text)})
        
       
        
  # First parsing method
# Run the Spider
process = CrawlerProcess()
process.crawl(RiaSrapper)
process.start()
#df = pd.DataFrame(df)
```

```python
import spacy
nlp_model = spacy.load('ru_core_news_md')
import string
```

```python
spacy_stopwords = spacy.lang.ru.stop_words.STOP_WORDS
new_stop_words = ['оренбург', 'оренбургский', 'год', 'область', 'район', 'город', 'ранее', 'теперь', 'областной', 'оренбуржье', 'рубль', 'ул.', 
                  'улица', 'дом', 'россия', 'оренбуржец', 'регион', '000', 'сегодня', 'время', 'центр', 'житель', 'человек', 'мужчина', 'женщина','...',
                 'ребенок']
spacy_stopwords.update(new_stop_words)
```

```python
spacy_stopwords
```

```python
def clean_str(sent):
    doc = nlp_model(sent)
    filter_str = [word.lemma_.lower() for word in doc if ((word.text not in string.punctuation) and (word.lemma_.lower() 
                                                                                                     not in spacy_stopwords) and (len(word.text) > 2))]
    filter_str = ' '.join(filter_str)
    return(filter_str)
```

```python
df['clean_content_news'] = df['content_news'].apply(clean_str)
```

```python
df.head()
```

```python
X = [x.split() for x in df['clean_content_news'].to_list()]
```

```python
X[:1]
```

```python
import gensim
from gensim import corpora

# Define the dictionary
dictionary = corpora.Dictionary(X)

# Define the corpus 
corpus = [dictionary.doc2bow(text) for text in X]

ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=10, id2word=dictionary, passes=5)

# Save the topics and top 5 words
topics = ldamodel.print_topics(num_words=10)
```

```python
for topic in topics:
    print(topic)
```

```python
import numpy as np
def get_topic_details(ldamodel, corpus):    
    topic_details = []
    for row in ldamodel[corpus]: 
        new_row = sorted(row, key=lambda x: (x[1]), reverse=True)    
        topic_details.append([new_row[0][0], new_row[0][1]])    
    topic_details_df = pd.DataFrame(topic_details, columns =['Dominant_Topic', 'Score'] )
    #topic_details_df.columns = ['Dominant_Topic', '% Score']
    return topic_details_df
# Add original text to topic details in a dataframe


contents = pd.DataFrame({'Original text':X})
topic_details = pd.concat([get_topic_details(ldamodel, corpus), contents], axis=1)

# Create flag for text highest associated with topic 3
#topic_details['flag'] = np.where((topic_details['Dominant_Topic'] == 5.0), 1, 0)
print(topic_details)
```
